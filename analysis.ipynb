# Data Cleaning

```python
print(f'Duplicate rows = {data.duplicated().sum()}')
```

```python
data.isna().sum()
```

```python
plt.figure(figsize=(6,6))
sb.heatmap(data.isnull())
plt.title("Graphical Overview of Missing Columns")
plt.show()
```

```python
data['retail_price'] = data['retail_price'].fillna(data['retail_price'].median())
data['discounted_price'] = data['discounted_price'].fillna(data['discounted_price'].median())
data['brand'] = data['brand'].fillna("Unknown")
data['product_rating'] = pd.to_numeric(data['product_rating'], errors='coerce')
data['overall_rating'] = pd.to_numeric(data['overall_rating'], errors='coerce')
```

```python
data.isna().sum()
```

```python
clean_data = data[data['product_rating'].notna()]
clean_data.isna().sum()
```

# Drop irrelevant Columns

```python
clean_data = clean_data.drop(columns=['uniq_id','crawl_timestamp','product_url','image', 'is_FK_Advantage_product', 'description','pid','product_specifications'])
```

# Feature Engineering 

```python
clean_data['discount'] = clean_data['retail_price'] - clean_data['discounted_price']
clean_data['discount_percentage'] = round((clean_data['discount'] / clean_data['retail_price']) * 100, 2)
clean_data[['retail_price', 'discounted_price', 'discount', 'discount_percentage']]
clean_data['main_category'] = clean_data['product_category_tree'].str.split(">>").str[0]
clean_data['main_category'] = clean_data['main_category'].str.replace(r'[\[\]\"]', '', regex=True)
clean_data['main_category'] = clean_data['main_category'].str.strip()
clean_data['rating_rounded'] = clean_data['product_rating'].round()
```

```python
clean_data.sample(5)
```

```python
plt.figure(figsize=(8,5))
sb.countplot(x='rating_rounded', data=clean_data)
plt.title("Product Rating Distribution")
plt.show()
```

#Obsevation 

1. The graph shows that the majority of products have a rating of 5, with the highest count among all rating categories.
2. Ratings are heavily concentrated in the 4–5 range, indicating that most products receive positive customer feedback.
3. Lower ratings (1–2) are relatively less frequent, suggesting that fewer products receive negative reviews.

```python
plt.figure(figsize=(8,5))

sb.histplot(clean_data['retail_price'], bins=50)

plt.title("Retail Price Distribution")
plt.xlabel("Price")
plt.show()
```

Observation 
1. Most products are concentrated in the low price range (₹0–₹2000).
2. The distribution is right-skewed, meaning high-priced products are fewer.
3. few extreme high-price outliers are present (above ₹50,000).

```python
plt.figure(figsize=(8,4))
sb.boxplot(x=clean_data['retail_price'])
plt.title("Retail Price Outliers")
plt.show()
```

Observation 
1. Most product prices are tightly clustered in the lower range, as shown by the compact box near the left.
2. A large number of outliers are present on the higher end, indicating many unusually expensive products.
3. The data is heavily skewed to the right, confirming the presence of extreme high-price values.

```python
plt.figure(figsize=(12,6))

top_brands = clean_data['brand'].value_counts().head(10)

sb.barplot(x=top_brands.values, y=top_brands.index)
plt.title("Top 10 Brands by Number of Products")
plt.xlabel("Count")
plt.ylabel("Brand")
plt.show()
```

observation

2. Among known brands, "Regular" and "Slim" have relatively higher product counts, but still far below "Unknown".

```python
plt.figure(figsize=(10,6))

top_categories = clean_data['main_category'].value_counts().head(10)

sb.barplot(y=top_categories.index, x=top_categories.values)

plt.title("Top 10 Categories by Number of Products")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

observation 
1. Clothing category dominates significantly, having the highest number of products by a large margin.
2. Watches, Computers, and Footwear are the next most common categories but far behind Clothing.
3. Categories like Mobiles & Accessories and Toys & School Supplies have the lowest product counts among the top 10.

Bivariate Analysis 

```python
plt.figure(figsize=(12,6))


top_brands = clean_data['brand'].value_counts().head(10).index
filtered_data = clean_data[clean_data['brand'].isin(top_brands)]

brand_discount = filtered_data.groupby('brand')['discount_percentage'].mean().sort_values(ascending=False)

sb.barplot(x=brand_discount.values, y=brand_discount.index)

plt.title("Average Discount % by Brand")
plt.ylabel("Brand")
plt.xlabel("Average Discount (%)")
plt.show()
```

Observation 
1. Black and Red brands offer the highest average discounts, indicating aggressive pricing or promotional strategies.
2. Regular and Slim brands also provide high discounts (~55%), making them competitive in pricing.
3. Clovia has the lowest average discount, suggesting relatively less price reduction compared to other brands.

```python
plt.figure(figsize=(12,6))

top_brands = clean_data['brand'].value_counts().head(10).index
filtered_data = clean_data[clean_data['brand'].isin(top_brands)]

brand_rating = filtered_data.groupby('brand')['product_rating'].mean().sort_values(ascending=False)

sb.barplot(x=brand_rating.index, y=brand_rating.values)
plt.title("Average Product Rating by Brand")
plt.ylabel("Average Rating")
plt.xlabel("Brand")
plt.show()
```

observation
1. Tenda and Netgear have the highest average ratings (~4.1–4.2), indicating strong customer satisfaction.
2. Most brands (Asus, Black, TP-LINK, Regular) have ratings close to 4, showing generally good product quality.
3. Clovia and Red have the lowest ratings (~3.2–3.3), suggesting comparatively lower customer satisfaction.

```python
plt.figure(figsize=(12,6))

# Top brands select karo
top_brands = clean_data['brand'].value_counts().head(10).index
filtered_data = clean_data[clean_data['brand'].isin(top_brands)]

# Boxplot
sb.boxplot(x='brand', y='retail_price', data=filtered_data)

plt.title("Retail Price Distribution by Brand")
plt.ylabel("Retail Price")
plt.xlabel("Brand")
plt.show()
```

```python
plt.figure(figsize=(12,6))

top_brands = clean_data['brand'].value_counts().head(5).index
top_categories = clean_data['main_category'].value_counts().head(5).index

filtered_data = clean_data[
    (clean_data['brand'].isin(top_brands)) &
    (clean_data['main_category'].isin(top_categories))
]

sb.countplot(x='main_category', hue='brand', data=filtered_data)


plt.title("Top Brands in Top Categories")
plt.legend(title="Brand")
plt.show()
```

observation
1. "Unknown" brand dominates across all major categories, especially in Clothing, Watches, and Footwear.
2. Clothing category has the highest brand diversity, with multiple brands like Slim, Clovia, and Regular contributing.
3. Other categories (Watches, Footwear, Computers) are heavily dominated by a single brand (Unknown), showing less brand variety.

```python
plt.figure(figsize=(8,6))

corr = clean_data[['retail_price','discounted_price','discount','discount_percentage','product_rating']].corr()

sb.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation")
plt.show()
```

Observation 
1. Retail price and discounted price are highly positively correlated (0.96), meaning higher-priced products still remain relatively expensive after discount.
2. Retail price and discount also show strong positive correlation (0.8), indicating expensive products tend to have higher absolute discounts.
3. Product rating has almost no correlation with price or discount variables, suggesting ratings are independent of pricing and discounts.

# Conclusion 
* The platform is primarily a budget-oriented marketplace with most products in low price ranges.
* Discount strategies are widely used, especially for high-priced products, but they don’t necessarily influence customer ratings.
* Customer satisfaction (ratings) is not dependent on pricing or discounts, suggesting other factors (quality, brand trust, etc.) matter more.
* A major data issue is the dominance of “Unknown” brands, which limits deeper brand-level insights.
* Clothing is the most competitive and saturated category, while others have less diversity.
